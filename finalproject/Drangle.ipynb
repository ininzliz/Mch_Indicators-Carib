```python
def abs_diff(df, col1, col2, col3=None):
    if col3 is None:
        df['abs_diff'] = df[col1].sub(df[col2]).abs()
    else:
        df['abs_diff'] = (df[col1].sub(df[col2]).abs() + df[col1].sub(df[col3]).abs() + df[col2].sub(df[col3]).abs()) / 3
    return df
```
